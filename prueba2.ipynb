# Analizando una Imagen

```python
import matplotlib.pyplot as plt 
img = plt.imread('beagle.jpg')
img.shape

img[10,10]
```

```python
# Hacemos una copia para que nos deje modificar los píxeles
img_copia = img.copy()

# Dibujamos la línea horizontal blanca
img_copia[150:160, :] = 255
plt.imshow(img_copia)
```

```python
# Dibujamos la línea vertical blanca cruzando la cabeza
img_copia[:, 550:560] = 255
plt.imshow(img_copia)
```

```python
# Recortamos la zona del fondo amarillo usando coordenadas
recorte = img_copia[0:150, 0:550]
plt.imshow(recorte)
```
